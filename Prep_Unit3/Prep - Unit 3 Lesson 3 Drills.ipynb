```python
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Gamma Distribution
gamma = np.random.gamma(10,3, 100)
plt.hist(gamma)
plt.axvline(gamma.mean(), color='c', linestyle='solid', linewidth=2)
plt.axvline(gamma.mean() + gamma.std(), color='c', linestyle='dashed', linewidth=2)
plt.axvline(gamma.mean() - gamma.std(), color='c', linestyle='dashed', linewidth=2)

plt.show()
```

```python
#Poisson Distribution
poisson = np.random.poisson(4, 100)
plt.hist(poisson)
plt.axvline(poisson.mean(), color='c', linestyle='solid', linewidth=2)
plt.axvline(poisson.mean() + poisson.std(), color='c', linestyle='dashed', linewidth=2)
plt.axvline(poisson.mean() - poisson.std(), color='c', linestyle='dashed', linewidth=2)

plt.show()
```

```python
#Binomial Distribution
binomial = np.random.binomial(50, 0.3, 100)
plt.hist(binomial)
plt.axvline(binomial.mean(), color='c', linestyle='solid', linewidth=2)
plt.axvline(binomial.mean() + binomial.std(), color='c', linestyle='dashed', linewidth=2)
plt.axvline(binomial.mean() - binomial.std(), color='c', linestyle='dashed', linewidth=2)

plt.show()
```

```python
#Bernoulli Distribution
bernoulli= np.random.binomial(1, .5, 100)
plt.hist(bernoulli)
plt.axvline(bernoulli.mean(), color='c', linestyle='solid', linewidth=2)
plt.axvline(bernoulli.mean() + bernoulli.std(), color='c', linestyle='dashed', linewidth=2)
plt.axvline(bernoulli.mean() - bernoulli.std(), color='c', linestyle='dashed', linewidth=2)


plt.show() 
```

```python
#Normal Distribution
normal = np.random.normal(10,5,100)
plt.hist(normal, bins=10, color='b')

plt.axvline(normal.mean(), color='c', linestyle='solid', linewidth=2)
plt.axvline(normal.mean() + normal.std(), color='c', linestyle='dashed', linewidth=2)
plt.axvline(normal.mean() - normal.std(), color='c', linestyle='dashed', linewidth=2)
plt.show
```

```python
#Conditional Distribution
cellphone = pd.DataFrame()
cellphone['calls'] = np.random.normal(50,10,100)
plt.hist(cellphone['calls'])
plt.axvline(cellphone['calls'].mean(), color='c', linestyle='solid', linewidth=2)
plt.axvline(cellphone['calls'].mean() + cellphone['calls'].std(), color='c', linestyle='dashed', linewidth=2)
plt.axvline(cellphone['calls'].mean() - cellphone['calls'].std(), color='c', linestyle='dashed', linewidth=2)
plt.show()
```

```python
var1 = np.random.normal(5, 0.5, 1000)
plt.hist(var1, bins=20, color='c')
plt.axvline(var1.mean(), color='k', linestyle='solid', linewidth=2)
plt.axvline(var1.mean() + var1.std(), color='k', linestyle='dashed', linewidth=2)
plt.axvline(var1.mean() - var1.std(), color='k', linestyle='dashed', linewidth=2)

var2 = np.random.normal(10, 1, 1000)
plt.hist(var2, bins=20, color='y')
plt.axvline(var2.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(var2.mean() + var2.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(var2.mean() - var2.std(), color='b', linestyle='dashed', linewidth=2)

plt.show()
```

```python
#Two normally distributed variable added together to create a third
var1 = np.random.normal(5, 0.5, 1000)
var2 = np.random.normal(10, 1, 1000)

var3 = var1 + var2
plt.hist(var3, bins=20, color='k')
plt.axvline(var3.mean(), color='y', linestyle='solid', linewidth=2)
plt.axvline(var3.mean() + var3.std(), color='y', linestyle='dashed', linewidth=2)
plt.axvline(var3.mean() - var3.std(), color='y', linestyle='dashed', linewidth=2)

plt.show()
```

```python
#Sample size of 100
pop1 = np.random.binomial(10, 0.2, 10000)
pop2 = np.random.binomial(10, 0.5, 10000)

sample1 = np.random.choice(pop1, 100, replace=True)
sample2 = np.random.choice(pop2, 100, replace=True)

print(sample1.mean())
print(sample2.mean())
print(sample1.std())
print(sample1.std())


```

```python
#Sample size of 1000
pop1 = np.random.binomial(10, 0.2, 10000)
pop2 = np.random.binomial(10, 0.5, 10000)

sample1 = np.random.choice(pop1, 1000, replace=True)
sample2 = np.random.choice(pop2, 1000, replace=True)

print(sample1.mean())
print(sample2.mean())
print(sample1.std())
print(sample1.std())

plt.hist(sample1, alpha=0.5, label='sample 1')
plt.hist(sample2, alpha=0.5, label='sample 2')
plt.legend(loc='upper right')
plt.show()
```

```python
#Sample size of 20
pop1 = np.random.binomial(10, 0.2, 10000)
pop2 = np.random.binomial(10, 0.5, 10000)

sample1 = np.random.choice(pop1, 20, replace=True)
sample2 = np.random.choice(pop2, 20, replace=True)

print(sample1.mean())
print(sample2.mean())
print(sample1.std())
print(sample1.std())

plt.hist(sample1, alpha=0.5, label='sample 1')
plt.hist(sample2, alpha=0.5, label='sample 2')
plt.legend(loc='upper right')
plt.show()
```

```python
#Sample size of 20 with change in pop1's probability value
pop1 = np.random.binomial(10, 0.3, 10000)
pop2 = np.random.binomial(10, 0.5, 10000)

sample1 = np.random.choice(pop1, 20, replace=True)
sample2 = np.random.choice(pop2, 20, replace=True)

print(sample1.mean())
print(sample2.mean())
print(sample1.std())
print(sample1.std())

diff=sample2.mean() - sample1.mean()
print (diff)

size = np.array([len(sample1), len(sample2)])
sd = np.array([sample1.std(), sample2.std()])

diff_se = (sum(sd ** 2 / size)) ** 0.5

print (diff/diff_se)

from scipy.stats import ttest_ind
print (ttest_ind(sample2, sample1, equal_var=False))


plt.hist(sample1, alpha=0.5, label='sample 1')
plt.hist(sample2, alpha=0.5, label='sample 2')
plt.legend(loc='upper right')
plt.show()
```

```python
#Sample size of 20 with another change in pop1's probability value
pop1 = np.random.binomial(10, 0.4, 10000)
pop2 = np.random.binomial(10, 0.5, 10000)

sample1 = np.random.choice(pop1, 20, replace=True)
sample2 = np.random.choice(pop2, 20, replace=True)

print(sample1.mean())
print(sample2.mean())
print(sample1.std())
print(sample1.std())

diff=sample2.mean() - sample1.mean()
print (diff)

size = np.array([len(sample1), len(sample2)])
sd = np.array([sample1.std(), sample2.std()])

diff_se = (sum(sd ** 2 / size)) ** 0.5

print (diff/diff_se)

from scipy.stats import ttest_ind
print (ttest_ind(sample2, sample1, equal_var=False))
```
